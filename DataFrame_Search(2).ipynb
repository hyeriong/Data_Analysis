# 데이터프레임 조회하기(2)

```python
import seaborn as sns
import pandas as pd
```

```python
# seaborn의 anagrams 데이터셋
anagrams = sns.load_dataset('anagrams')
```

```python
# anagrams 데이터셋의 행렬 개수
anagrams.shape
```

## 행 조회 (+ 열 조회)
### iloc
- integer location의 약어
- 컴퓨터가 읽을 수 있는 indexing 값으로 데이터프레임의 행이나 컬럼에 접근하는 것이 특징
- 테이블명.iloc[행 인덱스, 열 인덱스]


```python
# 첫 번째 행 접근하기
anagrams.iloc[0]
```

```python
# 두 번째 행 접근하기
anagrams.iloc[1]
```

```python
# 마지막 행 접근
anagrams.iloc[-1]
```

```python
# 첫 번째 열 조회하기
anagrams.iloc[:, 0]
```

```python
# 두 번째 열 조회하기
anagrams.iloc[:, 1].head()
```

```python
# 두 번째 열의 세 번째 행 정보 조회
anagrams.iloc[2, 1]
```

```python
# 슬라이싱을 이용해 구간 정보 조회하기
anagrams.iloc[:3, :5]  # 데이터프레임으로 출력되는 것이 특징
```

```python
# 짝수 번째에 위치한 행들만 조회하기
anagrams.iloc[::2, :]
```

```python
# 2~4번 열의 6~10번 행 조회하기
anagrams.iloc[6:11, 1:4]
```

```python
# 만약 컬럼명을 인덱싱 값으로 넣는다면?
anagrams.iloc[8, 'attnr']   # 오류 발생
```

```python
# 이에 대한 해결 방법은?
anagrams.iloc[8]['attnr']
```

```python
# 람다 함수 사용하여 조회하기
anagrams.iloc[lambda x: x.index == 13]
```
