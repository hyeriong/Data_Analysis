## sort_values()
- 값을 기준으로 레이블을 정렬하는 메서드
- df.sort_values(by, axis = 0, ascending = True, inplace = False, kind = 'quicksort', na_position = 'last', ignore_index = False, key = None)
  - by: 정렬 기준이 될 레이블
  - axis: {0:index/ 1:columns} 정렬할 레이블, 0이면 행, 1이면 열
  - inplace: 원본을 대체할지 여부, True이면 원본 대체
  - kind: 알고리즘 모드
  - na_position: 결측치 위치, 기본값은 last로 정렬시 맨 뒤에 위치
  - ignore_index: 인덱스 무시 여부, True이면 인덱스의 순서와 상관없이 0,1,2,...로 정해짐
  - key: 이 인수를 통해 정렬방식으로 함수 사용

```python
flights.sort_values('passengers')  # 디폴트: 오름차순
```

```python
flights.sort_values('passengers', ascending = True)  # 즉, 디폴트는 ascending = Ture임
```

```python
flights.sort_values('passengers', ascending = False)  # ascending = False하면 내림차순
```
