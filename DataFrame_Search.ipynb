## 행 조회(+ 열 조회)

### loc
- location의 약자
- 인덱스 네임으로 조회하는 방식
- 데이터프레임의 행렬 라벨 혹은 boolean array을 통해 데이터 조회
- 테이블명.loc[행 인덱싱 값, 열 인덱싱 값]

```python
iris.head()  
```

```python
# 1. loc
# 첫 번째 행 추출하기
iris.loc[0]
```

```python
# 이때도 마찬가지로 대괄호 두 번 쓰면 데이터프레임 형식으로 출력된다.
iris.loc[[0]]
```

```python
# 두 번째 행의 species 컬럼 추출
iris.loc[1, 'species']
```

```python
# 행 인덱싱 값 부분에 :을 쓰면 species 컬럼의 모든 값을 가져온다(슬라이싱)
iris.loc[:, 'species']
```

```python
# sepal_length ~ petal_length컬럼의 전체 값을 가져오기
iris.loc[:, :'petal_length']
```

```python
# 위에 출력된 데이터프레임에서 3번째 줄까지만 가져오기
iris.loc[:3, :'petal_length']
```

```python
# 조건을 붙여서(conditioning) 추출할 수도 있다.
iris.loc[(iris['sepal_length'] > 7) & (iris['species'] == 'virginica')]
```

```python
# 조건을 붙여서 추출한 데이터의 특정 컬럼 값 추출
iris.loc[(iris['sepal_length'] > 6) & (iris['petal_length'] > 6.3), 'petal_width':'species']
```
