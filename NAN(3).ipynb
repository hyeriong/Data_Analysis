- Per컬럼과 ROE컬럼, Dividend컬럼에서 결측치 확인됨

```python
# 컬럼이 많은 관계로 결측치 데이터만 살펴보겠음
sample = stock_market[['PER', 'ROE', 'Dividend']]
```

```python
sample
```

## 결측값 제거 방법
- 1) 결측값이 있는 행 제거
- 2) 결측값이 있는 열 제거

```python
# 결측값이 있는 행 일괄제거
sample.dropna()
```

- 단점: 하나의 결측값이라도 있으면, 모든 컬럼에서 행이 삭제됨. -> 결측값이 많을 경우, 데이터셋이 확연히 줄어들 수 있음.

```python
# 결측값이 하나뿐인 행 살펴보기
sample[sample['PER'].isnull() | sample['ROE'].isnull()]
```

```python
sample.iloc[223]  # 223행은 결측값이 존재하는 행
```

```python
# 특정컬럼(PER, ROE)에서만 결측값이 있는 행 제거하기
sample.dropna(subset = ['PER','ROE']).iloc[223]  # 223행에서 결측값이 사라진 것을 볼 수 있음.
```

```python
# 특정 열 제거(결측치가 많은 열 제거)
sample.drop(['Dividend'], axis = 1)
```

## 결측치 대체 방법
- fillna(method = '  ')
  - ffill: 결측치가 나타나기 전의 값으로 결측치 변경
  - bfill: 결측값이 나타난 이후 첫 번째 값으로 앞쪽의 결측치를 변경

```python
# 결측치 대체 전의 Dividend 평균
sample.Dividend.mean()
```

```python
# 변수명 의미: ffill로 결측값을 채운 sample
ff_sample = sample.fillna(method = 'ffill')['Dividend']
```

```python
# ff_sample의 평균구하기
ff_sample.mean()
```

- 결측치 대체 이전과 대략 0.16 정도의 차이 발생

```python
# bfill로 결측값 대체해보기
bf_sample = sample.fillna(method = 'bfill')['Dividend']
```

```python
# bf_sample의 평균 구하기
bf_sample.mean()
```

- 결측치 대체 이전과 대략 0.004 정도의 차이 발생
