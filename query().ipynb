```python
import io
import pandas as pd
```

```python
wine = pd.read_csv(io.BytesIO(uploaded['wine.csv']))
```

```python
wine.info()
```

```python
# 컬럼명 공백 제거
wine.columns = wine.columns.str.replace(' ', '_')
```

```python
wine.columns
```

컬럼명의 의미 <br>
- fixed acidity : 고정산
- volatile acidity : 휘발산
- critric acid : 구연산
- residual sugar : 잔당
- chlorides : 염화물
- free sulfur dioxide : 유리 이산화황
- total sulfur dioxide : 총 이산화황
- density : 밀도
- pH (산성도)
- sulphates : 황산염
- alcohol : 알코올
- quality : 등급

## 특징
- 정제된 데이터에서 특정 조건을 만족하는 결과를 추출하기 위한 함수
- 가독성과 편의성이 뛰어나다는 장점
- 비교연산, in연산, 논리 연산, 외부 변수 및 참조 연산, 인덱스 검색, 문자열 부분 검색 등을 이용해 표현 가능

```python
# 잔당이 2.0 이하인 레드와인 필터링
wine.query('type == "red" and residual_sugar <= 2.0').head()
```

```python
# 고정산이 10보다 크고 휘발산이 0.600 이하인 와인
wine.query('fixed_acidity > 10 and volatile_acidity <= 0.600').head()
```

```python
# 5등급인 레드와인
wine.query('type in "red" and quality == 5')[['type', 'quality']].head()
```

```python
# 외부변수 또는 함수의 참조 연산 이용하기
pH_wanted = 3.12

wine.query('pH == @pH_wanted')[['type', 'pH']].head()
```

```python
# 1000번째 행 검색
wine.query('index == 1000')
```

```python
# wh로 시작하는 타입의 와인 찾기
wine.query('type.str.startswith("wh")').head()
```

```python
# d로 끝나는 타입의 와인 찾기
wine.query('type.str.endswith("d")').head()
```
