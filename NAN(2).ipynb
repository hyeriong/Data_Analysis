# 결측치 대체

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
```

```python
bookings = pd.read_csv(io.BytesIO(uploaded['Bookings.csv']))
```

```python
bookings
```

```python
bookings.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
# 데이터 내보내기
bookings.to_csv('/content/bookings.csv', index = False)
```

```python
bookings.head()
```

## 결측값을 0으로 대체하기
- 아래로 내려갈수록 결측값이 많아진다. --> 즉, 아래쪽 데이터는 고객 방문율이 낮음을 의미한다.
- 따라서 호텔 방문에 대한 결측값을 '리뷰 없음'으로 판단하고, 이를 0으로 채운다.
- 다만, Review 컬럼과 같은 경우 문자열이므로, 이에 해당하는 value를 하나 생성한다.

```python
bookings.info()
```

- Rating을 제외하고는 전부 문자열 value임.

```python
bookings[['Total_Review', 'Rating']] = bookings[['Total_Review', 'Rating']].fillna(0)
```

```python
bookings.tail()
```

## 결측치에 새로운 value 할당하기
- Review 컬럼은 문자열이므로 새로운 value를 생성하여 부여한다

```python
bookings['Review'] = bookings['Review'].fillna('None')
```

```python
bookings.tail()
```
